```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
df = pd.read_csv('./dataset/trainingset.csv', header=None)
df.head(10)
```

```python
y = df.iloc[2500:2800, 6].values  # 预测标签向量
y1 = df.iloc[2500:2800, 1].values
x = df.iloc[2900:3200, 6].values  # 输入特征向量
x1 = df.iloc[2900:3200, 1].values
# 使用散点图可视化样本
a1 = plt.figure(1)
plt.scatter(x, x1, color='red', marker='o', label='no')
plt.scatter(y, y1, color='blue', marker='x', label='yes')
plt.legend(loc='upper left')
a1.show
```

```python
for q in range(1,16):
    y = df.iloc[2500:2800, q].values  # 预测标签向量
    y1 = df.iloc[2500:2800, 11].values
    x = df.iloc[2900:3200, q].values  # 输入特征向量
    x1 = df.iloc[2900:3200, 11].values
    # 使用散点图可视化样本
    a1 = plt.figure(q)
    plt.scatter(x, x1, color='red', marker='o', label='no')
    plt.scatter(y, y1, color='blue', marker='x', label='yes')
    plt.legend(loc='upper left')
    a1.show
   



```
