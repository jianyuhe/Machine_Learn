```python
contintable = pd.DataFrame({'age':nd['age'].values, 'fnlwgt':nd['fnlwgt'].values, 'education-num':nd['education-num'].values,
                           'capital-gain':nd['capital-gain'].values, 'capital-loss':nd['capital-loss'].values,
                          'hours-per-week':nd['hours-per-week'].values})
cvalue = contintable.isnull().sum().values
missv = []
for i in cvalue:
    missv.append(i/len(contintable['fnlwgt'])*100)

missv
```

```python
card = [contintable['age'].nunique(),contintable['fnlwgt'].nunique(),contintable['education-num'].nunique(),
        contintable['capital-gain'].nunique(),
        contintable['capital-loss'].nunique(),
       contintable['hours-per-week'].nunique()]
```

```python

contreport = pd.DataFrame({'FEATURES':descri,
                   'COUNT':des.iloc[0].values,
                   'MISS %': missv,
                   'CARD':card,
                  'MIN':des.iloc[3].values,
                   '1st QRT':des.iloc[4].values,
                  'MEAN' : des.iloc[1].values,
                  'MEDIUM':des.iloc[5].values,
                  '3rd QRT': des.iloc[6].values,
                  'MAX': des.iloc[7].values,
                  'STD':des.iloc[2].values})
```

```python
contreport.to_csv('./data/studentnumberCONT.csv',sep = ',', index=False)
contreport
```

```python
catetable = pd.DataFrame({'workclass':nd['workclass'].values, 'education':nd['education'].values, 'marital-status':nd['marital-status'].values,
                           'occupation':nd['occupation'].values, 'relationship':nd['relationship'].values,
                          'race':nd['race'].values,
                        'sex':nd['sex'].values,
                         'native-country':nd['native-country'].values})
catetable
```

```python
catetable.count().values
```

```python
cateheader = ['workclass','education','marital-status','occupation','relationship','race','sex','native-country']
```

```python
catevalue = catetable.isnull().sum().values
catemiss = []
for i in catevalue:
    catemiss.append(i/len(catetable['workclass'])*100)
    
catemiss
```

```python
catecard = []
for i in cateheader:
    catecard.append(catetable[i].nunique())
catecard
```

```python
maxindex =[]
maxvalue=[]
percen =[]
countv = []
for i in cateheader:
    r = catetable[i].value_counts().sort_values(ascending=False)
    maxindex.append(r.index[0])
    maxvalue.append(r.values[0])
    percen.append(r.values[0]/len(catetable[i])*100)
    countv.append(len(catetable[i]))
percen
```

```python
maxindex2 =[]
maxvalue2=[]
percen2 =[]
for i in cateheader:
    r = catetable[i].value_counts().sort_values(ascending=False)
    maxindex2.append(r.index[1])
    maxvalue2.append(r.values[1])
    percen2.append(r.values[1]/catetable[i].count()*100)
percen2
```

```python
catereport = pd.DataFrame({'FEATURES':cateheader,
                   'COUNT':countv,
                   'MISS %': catemiss,
                   'CARD':catecard,
                  'MODE':maxindex,
                   'MODE FREQ':maxvalue,
                  'MODE %' : percen,
                  '2nd MODE':maxindex2,
                  '2nd MODE FREQ': maxvalue2,
                  '2nd MODE %': percen2})
```

```python

catereport.to_csv('./data/studentnumberCAT.csv',sep = ',', index=False)
catereport
```
